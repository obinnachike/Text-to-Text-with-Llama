```python
from llama_index.core import Settings
from llama_index.core.node_parser import SentenceSplitter
from llama_index.llms.gemini import Gemini


```

```python
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-base-en")

llm = Gemini(model="gemini-2.5-pro")
Settings.llm = llm

Settings.embed_model = embed_model

Settings.node_parser = SentenceSplitter(chunk_size=800, chunk_overlap=20)
```

```python
index = VectorStoreIndex.from_documents(documents)
```

## Storing and Loading the Index

```python
index.storage_context.persist()
```

## Q/A

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What is Haloperidol?")
response
```
